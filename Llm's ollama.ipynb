```python
import ollama

stream = ollama.chat(
        model = MODEL,
    messages=message,
    stream=True
)
```

```python
for chunk in stream:
    print(chunk['message']['content'],end='',flush=True)
```

```python
law_system_prompt = """you are a ai assitant related Indian Laws, your task is to give a good verified 
                        information about anything related to indian law or realted cases, don't 
                        provide any unverified information, if no information is provided give it as no 
                        furture information is available or something like that, max no of tokens should be 1000, 
                        and 2-3 paragraphs, make sure to complete the sentance by end"""
stream = ollama.chat(
    model='hf.co/varma007ut/Indian_Legal_Assitant:latest',
    messages=[{'role':'system', 'content': law_system_prompt},{'role': 'user', 'content':'What are the fundamental rights in the Indian Constitution?'}],
    stream=True,
    options={
        'num_predict': 1000  # Sets the maximum length to 150 tokens
    }
)
for chunk in stream:
    print(chunk['message']['content'],end='',flush=True)
```

```python


ollama_model = OpenAI(base_url=ollama_url,api_key='ollama')
response = ollama_model.chat.completions.create(model=MODEL,messages=message)
```

```python
from IPython.display import Markdown, display

display(Markdown(response.choices[0].message.content))
```
