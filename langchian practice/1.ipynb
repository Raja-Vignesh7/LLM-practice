```python
from langchain_google_genai.llms import GoogleGenerativeAI
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser

llm = GoogleGenerativeAI(model="gemini-2.5-flash")
prompt = PromptTemplate.from_template("You are a personal assistant. Answer the following question: {question}")
output_parser = StrOutputParser()

print(prompt)
```

```python
print(dir(prompt))
print(prompt.get_input_schema)
```

```python
chain = prompt | llm | output_parser
print(chain.invoke({"question": "explain difference between Ai agents and Agentic AIs?"}))
```

```python
print(dir(chain))
```

```python
# from langchain_community.chat_models import ChatHuggingFace
from langchain_huggingface.llms import HuggingFacePipeline
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline
import torch

model_id = "meta-llama/Llama-3.2-1B"

device = "cuda" if torch.cuda.is_available() else "cpu"
```

```python
model = AutoModelForCausalLM.from_pretrained(model_id,dtype='auto').to(device)
tokenizer = AutoTokenizer.from_pretrained(model_id)
```

```python
llama_pipeline = pipeline('text-generation', model=model, tokenizer=tokenizer, device=device,max_new_tokens=100)
```

```python
hf = HuggingFacePipeline(pipeline=llama_pipeline)
```
